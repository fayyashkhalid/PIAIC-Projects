```python
import random
from collections import Counter

Words_list = '''apple banana mango strawberry orange grape pineapple apricot lemon coconut watermelon cherry papaya berry peach lychee muskmelon'''

Words_list = Words_list.split(' ')
word = random.choice(Words_list)

if __name__ == '__main__':
    print('Guess the word! HINT: word is a name of a fruit')

    for i in word:
        print('_', end=' ')
    print()

    playing = True
    letter_Guess = ''
    tries = len(word) + 2
    correct = 0
    flag = 0
    try:
        while (tries != 0) and flag == 0:
            print()
            tries -= 1

            try:
                guess = str(input('Enter a letter to guess (e.g. a to z): '))
            except:
                print('Enter only a letter!')
                continue

            if not guess.isalpha():
                print('Please Enter only LETTER (e.g. a to z)')
                continue
            elif len(guess) > 1:
                print('Enter only a Single letter')
                continue
            elif guess in letter_Guess:
                print('You have already guessed that letter')
                continue
            if guess in word:
                k = word.count(guess)
                for _ in range(k):
                    letter_Guess += guess
            for char in word:
                if char in letter_Guess and (Counter(letter_Guess) != Counter(word)):
                    print(char, end=' ')
                    correct += 1
                elif (Counter(letter_Guess) == Counter(word)):
                    print("The word is: ", end=' ')
                    print(word)
                    flag = 1
                    print('Congratulations, You won!')
                    break
                    break
                else:
                    print('_', end=' ')
        if tries <= 0 and (Counter(letter_Guess) != Counter(word)):
            print()
            print('You lost! Try again..')
            print('The word was {}'.format(word))

    except KeyboardInterrupt:
        print()
        print('Bye! Try again.')
        exit()
```
