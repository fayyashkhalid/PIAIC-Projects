```python
import random
def user_guess():
    number = random.randint(1,50)
    guesses_availble = 10
    print("Welcome to User Guess the Number Game")
    print("Please choice Number between 1 to 50")
    while guesses_availble > 0:
        print(f"You have {guesses_availble} Guessess Left")
        guess = int(input("Please Guess a Number: "))
        if(guess < number):
            print("Guess Number Too Low, Chioce another Number: ")
        elif(guess > number):
            print("Guess Number Too High, Chioce another Number: ")
        else:
            print(f"Congratulation! You Choice the Correct Number in {10 - guesses_availble+1}, tries")
            return
        guesses_availble -= 1
    print(f"Oops Your Tries is end. The Number is {number}")
user_guess()
```
