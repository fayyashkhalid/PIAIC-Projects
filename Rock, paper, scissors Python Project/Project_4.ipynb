```python
import random
option = ("rock", "paper", "scissors")
played = True
while played:
  user = None
  computer = random.choice(option)
  while user not in option:
    user = input("Enter Your Choice: (e.g. 0 for Rock, 1 for Paper, 2 for Scissors): ")
  # if user not in option:
  #   print("Please Enter Valid Choice")
  # else:
    print(f"User Choice: {user}")
    print(f"Computer Choice: {computer}")

    if (computer == user):
      print("Its a draw") #done
    elif (computer == "rock" and user == "scissors"):
      print("Opps, You lose") #done
    elif (computer == "paper" and user == "rock"):
      print("Opps, You lose") #done
    elif (computer == "scissors" and user == "paper"):
      print("Opps, You lose") #done

    elif (user == "rock" and computer == "scissors"):
      print("Yahooo, You Won") #done
    elif (user == "paper" and computer == "rock"):
      print("Yahooo, You Won") #done
    elif (user == "scissors" and computer == "paper"):
      print("Yahooo, You Won") #done

    if not input("Play again? (y/n): ").lower() == "y":
      played = False
print(f"Thanks for playing!")
```
